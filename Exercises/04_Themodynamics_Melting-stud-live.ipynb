### Melting of a pure substance
1) Set the initial enthaly and the enthalpy of at onset of fusion
2) Define the conditions for melting (complete `if` and `elseif` statement) 
3) Compute temperature (`T`) and the porosity ($\phi$) for all conditions (before, during, after melting)

```julia
# Physics
T_fusion = 773.0  # K
Q_latent = 400e3  # J/kg
T_ini    = 200.0
Q̇        = 1e-7
cp       = 1000.0  # J/kg/K

# Numerics
t_tot    = 0.5*My
nt       = 1000
Δt       = t_tot/nt
t        = 0.0

# Initial values
h_fusion = cp*T_fusion  # Enthalpy et fusion temperature
h        = cp*T_ini     # Initial enthalpy

# Allocate vectors
T_vec    = T_ini*ones(nt) 
h_vec    = h*ones(nt) 
ϕ_vec    = 0.0*ones(nt) 
t_vec    = t*ones(nt)

# Loop over time
for it=2:nt

    t = t + Δt

    h = h + Δt*Q̇

    if h<h_fusion # 100 % solid
        T = h/cp
        ϕ = 0.0 
    elseif h>h_fusion+Q_latent # 100% liquid
        T = (h - Q_latent)/cp
        ϕ = 1.0 
    else
        T = h_fusion/cp
        ϕ = (h - h_fusion)/Q_latent      
    end

    # Store in vectors for plotting
    h_vec[it] = h
    T_vec[it] = T
    ϕ_vec[it] = ϕ
    t_vec[it] = t
end

# Visualise
p1 = plot(h_vec./1e3, T_vec, ylabel="T [K]", xlabel="h [kJ/kg]")
p2 = plot(h_vec./1e3, ϕ_vec, ylabel="ϕ [-]", xlabel="h [kJ/kg]")
plot(p1, p2)
```
